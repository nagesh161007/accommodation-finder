## Accommodation Finder

#### Import web scraped data from csv to data frame

```python
import pandas as pd
from IPython.core.display import HTML
accomodationData = pd.read_csv('accommodation.csv')
df = pd.DataFrame(accomodationData)
display(HTML(df.to_html()))


# convert 
```

```python
print(df.info())
```

```python
print(df.dtypes)
```

#### Cleaning data

```python
# convert rent from dollars to float

df['rent'] = df['rent'].str.split('-').str[0].str.strip('$')
df['rent'] = df['rent'].str.split('/').str[0].str.replace(',','')

df['rent'] = df['rent'].astype('float')

assert df['rent'].dtype == 'float'

#HTML(df[['rent']].to_html())
```

```python
# cleaning image urls

df['image'] = df['image'].astype('str')
df['image'] = df['image'].str.strip('nan')
df['image'] = df['image'].str.strip('|')
HTML(df[['image']].to_html())
```

```python
# cleaning description

df['description'] = df['description'].astype('str')
df['description'] = df['description'].fillna(' ')
HTML(df[['description']].to_html())
```

```python
# cleaning address

df['address'] = df['address'].astype('str')
df['address'] = df['address'].str.replace('nan','')
HTML(df[['address']].to_html())
```

```python
# create bed and bath 

df['bed'] = df['roomtype'].str.split(' ').str[0]

df['bed'] = df['bed'].astype('int')

HTML(df[['bed']].to_html())
```

```python
df['bath'] =  df['roomtype'].str.split('•').str[1]
df['bath'] =  df['bath'].str.split(' ').str[1];

df['bath'] = df['bath'].astype('float')

HTML(df[['bath']].to_html())
```

```python
df['availability']  = df['availability'].str.lower().str.replace('available',' ')
HTML(df[['availability']].to_html())
```

```python
from datetime import date

df.loc[df['availability'].str.contains('now'), 'availability']  = date.today().strftime('%B %d')

HTML(df[['availability']].to_html())
```

```python
df['locality'] = df['locality'].str.replace('\n',' ')
df['locality'].fillna(' ', inplace = True)
HTML(df[['locality']].to_html())
```

## 

```python
df['male'].fillna(value=int(round(df['male'].mean())), inplace=True)
df['female'].fillna(value=int(round(df['female'].mean())), inplace=True)
```

```python
# HTML(df[['bath']].to_html())
df['bath'].fillna(value=int(round(df['bath'].mean())), inplace=True)
```

```python
df['locality'] = df['locality'].astype('category')
df.info()
```

```python
df['availability'] = df['availability'].str.strip()

df['availability'] = pd.to_datetime(df['availability'], format="%B %d")
```

### Temporary Accommodation

```python
import pandas as pd
from IPython.core.display import HTML
tempAccomodationData = pd.read_csv('TemporaryAccomodation.csv')
tpdf = pd.DataFrame(tempAccomodationData)
display(HTML(tpdf.to_html()))
```

```python
tpdf['rent'].fillna(value=int(round(tpdf['rent'].mean())), inplace=True)
display(HTML(tpdf.to_html()))
```

```python
tpdf['gender'].fillna(value='both', inplace=True)
display(HTML(tpdf.to_html()))
```

```python
tpdf['description'].fillna(value='', inplace=True)
display(HTML(tpdf.to_html()))
```

```python
tpdf['spot'].fillna(value='SHARED', inplace=True)
display(HTML(tpdf.to_html()))
```

```python
tpdf['name'].fillna(value='', inplace=True)
tpdf['number'].fillna(value='', inplace=True)
display(HTML(tpdf.to_html()))
```
